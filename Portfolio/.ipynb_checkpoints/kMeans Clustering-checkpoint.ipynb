```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ucimlrepo import fetch_ucirepo
```

# Reading and Exploring the Dataset

```python
# fetch dataset
heart_disease = fetch_ucirepo(id=45) 
  
# data (as pandas dataframes) 
X = heart_disease.data.features 
y = heart_disease.data.targets 

heart_disease_df = pd.DataFrame()
heart_disease_df = pd.concat([X,y],axis=1)
```

```python
heart_disease_df.head()
```

```python
heart_disease_df.info()
```

```python
heart_disease_df.dropna(axis=0,how='any',inplace=True)
```

```python
heart_disease_df.describe()
```

```python
heart_disease_df.hist(bins=30,figsize=[20,20])
plt.show()
```

```python
heart_disease_df['num'].value_counts()
```

```python
heart_disease_df['diagnosis'] = heart_disease_df['num'].apply(lambda x: 0 if x==0 else 1)
```

```python
heart_disease_df['diagnosis'].value_counts()
```

```python
plt.figure(figsize=[14,10])
cmap = sns.diverging_palette(260, 10, as_cmap=True)
sns.heatmap(heart_disease_df.corr(),annot=True,cmap=cmap,vmin=-1.2,vmax=+1.2)
plt.title("Heart Disease Correlation Matrix")
plt.show()
```

```python
X = heart_disease_df.drop(['num','diagnosis'],axis=1)
```

```python
X.columns
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
kmeans_model = KMeans(n_clusters=2,max_iter=300,n_init=20,random_state=42)
kmeans_model.fit(X)
```

```python
kmeans_model.cluster_centers_
```

```python
cluster_means_df = pd.DataFrame(kmeans_model.cluster_centers_,columns=X.columns)
cluster_means_df
```

```python
heart_disease_df[heart_disease_df['diagnosis']==0].describe()
```

```python
heart_disease_df[heart_disease_df['diagnosis']==1].describe()
```

```python
cluster_analysis_df = pd.DataFrame()
cluster_analysis_df['K-Means cluster centroid 1'] = cluster_means_df.iloc[0]
cluster_analysis_df['K-Means cluster centroid 2'] = cluster_means_df.iloc[1]

cluster_analysis_df['Mean of corresponding features(num=0)'] = heart_disease_df[heart_disease_df['diagnosis']==0].mean()
cluster_analysis_df['Mean of corresponding features(num=1)'] = heart_disease_df[heart_disease_df['diagnosis']==1].mean()
```

```python
cluster_analysis_df
```

* Based on similarity between cluster 1 and the data points corresponding to diagnosis = 0 we can allot cluster 1 to label 0
* Similarly, Based on similarity between cluster 2 and the data points corresponding to diagnosis = 1 we can allot cluster 2 to label 1

```python
y_pred = kmeans_model.labels_
y = heart_disease_df['diagnosis']
```

```python
print("Accuracy of the model:",round(accuracy_score(y,y_pred),2))
```

```python
conf_matrix = confusion_matrix(y,y_pred)
fig,ax = plt.subplots(figsize=[5,5])
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix)
disp.plot(cmap=plt.cm.PuBu,ax=ax)
ax.set_xlabel("Predicted Value")
ax.set_ylabel("Actual Value")
plt.show()
print(classification_report(y,y_pred))
```
